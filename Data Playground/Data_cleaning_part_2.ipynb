```python
import numpy as np
from sklearn.preprocessing import StandardScaler

# Create sample data (age and income)
age = np.array([30, 40, 50, 60, 70])
income = np.array([50000, 60000, 70000, 80000, 90000])
```

```python
# Create sample data (age and income)
age = np.array([30, 40, 50, 60, 70])
income = np.array([50000, 60000, 70000, 80000, 90000])

# Combine features into a 2D array
data = np.column_stack((age, income))
```

```python
print(data)
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

print(data)
print(scaled_data)
```

So basically we used the standard scaler function to standardise the data.
Now we perform the PCA on the scaled data.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(scaled_data)
```

```python


# Step 3: Access Transformed Data
principalComponents = pca.transform(scaled_data)
principalComponents
```

```python
import matplotlib.pyplot as plt

# Plot principal components
plt.figure(figsize=(8, 6))
plt.scatter(principalComponents[:, 0], principalComponents[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Plot')
plt.grid(True)
plt.show()
```
