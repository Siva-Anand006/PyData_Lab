```python
Features = ['Pregnancies',
 'Glucose',
 'BloodPressure',
 'SkinThickness',
 'Insulin',
 'BMI',
 'DiabetesPedigreeFunction',
 'Age',]
# Separating out the features
x = diabetes_dataset.loc[:, Features].values

#Separating the outcomes
y = diabetes_dataset.loc[:,['Outcome']].values #y = df.loc[:,['target']].values
```

```python
print(x)
print(y)
```

```python
scalar = StandardScaler()
Standardised_data = scalar.fit_transform(x)
```

```python
print(Standardised_data)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

principalComponents = pca.fit_transform(Standardised_data)

principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```
